```python
import pandas as pd

# Load the vehicles dataset from the CSV file
df_vehicles = pd.read_csv('Motor_Vehicle_Collisions_-_Vehicles_20240412.csv')

# Check missing values in each column
print("Missing values in each column:")
print(df_vehicles.isnull().sum())
```

```python
# Handle missing values
df_vehicles['VEHICLE_MAKE'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_MODEL'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_YEAR'].fillna(df_vehicles['VEHICLE_YEAR'].median(), inplace=True)
df_vehicles['DRIVER_SEX'].fillna('Unknown', inplace=True)
df_vehicles['DRIVER_LICENSE_STATUS'].fillna('Unknown', inplace=True)
df_vehicles['DRIVER_LICENSE_JURISDICTION'].fillna('Unknown', inplace=True)
df_vehicles['PRE_CRASH'].fillna('Unknown', inplace=True)
df_vehicles['POINT_OF_IMPACT'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_DAMAGE'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_DAMAGE_1'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_DAMAGE_2'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_DAMAGE_3'].fillna('Unknown', inplace=True)
df_vehicles['PUBLIC_PROPERTY_DAMAGE'].fillna('Unknown', inplace=True)
df_vehicles['PUBLIC_PROPERTY_DAMAGE_TYPE'].fillna('Unknown', inplace=True)
df_vehicles['CONTRIBUTING_FACTOR_1'].fillna('Unknown', inplace=True)
df_vehicles['CONTRIBUTING_FACTOR_2'].fillna('Unknown', inplace=True)

# Convert data types
df_vehicles['CRASH_DATE'] = pd.to_datetime(df_vehicles['CRASH_DATE'])
df_vehicles['CRASH_TIME'] = pd.to_datetime(df_vehicles['CRASH_TIME'], format='%H:%M', errors='coerce').dt.time
df_vehicles['VEHICLE_YEAR'] = df_vehicles['VEHICLE_YEAR'].astype('Int64')  # Use 'Int64' to allow nullable integers
df_vehicles['VEHICLE_OCCUPANTS'] = df_vehicles['VEHICLE_OCCUPANTS'].astype('Int64')

# Standardize categorical values
df_vehicles['DRIVER_SEX'] = df_vehicles['DRIVER_SEX'].str.title()
df_vehicles['DRIVER_LICENSE_STATUS'] = df_vehicles['DRIVER_LICENSE_STATUS'].str.title()

# Save the cleaned dataset to a new CSV file
df_vehicles.to_csv('cleaned_vehicles_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned vehicles dataset from the CSV file
df_vehicles = pd.read_csv('cleaned_vehicles_data.csv')

# Handle missing values
df_vehicles['TRAVEL_DIRECTION'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_OCCUPANTS'].fillna(df_vehicles['VEHICLE_OCCUPANTS'].median(), inplace=True)
df_vehicles['STATE_REGISTRATION'].fillna('Unknown', inplace=True)
df_vehicles['VEHICLE_TYPE'].fillna('Unknown', inplace=True)

# Standardize inconsistent values in 'VEHICLE_TYPE'
df_vehicles['VEHICLE_TYPE'] = df_vehicles['VEHICLE_TYPE'].str.upper()
df_vehicles['VEHICLE_TYPE'].replace(['TLC', 'DOT', 'UNKNOWN VE'], 'OTHER', inplace=True)

# Handle outliers in 'VEHICLE_YEAR' and 'VEHICLE_OCCUPANTS'
df_vehicles = df_vehicles[(df_vehicles['VEHICLE_YEAR'] >= 1900) & (df_vehicles['VEHICLE_YEAR'] <= 2023)]
df_vehicles = df_vehicles[df_vehicles['VEHICLE_OCCUPANTS'] <= 10]

# Drop the 'DRIVER_SEX' column as it doesn't add value to the analysis
df_vehicles.drop('DRIVER_SEX', axis=1, inplace=True)

# Save the updated cleaned dataset to a new CSV file
df_vehicles.to_csv('updated_cleaned_vehicles_data.csv', index=False)

# Plot the distribution of top vehicle types
top_vehicle_types = df_vehicles['VEHICLE_TYPE'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.countplot(x='VEHICLE_TYPE', data=df_vehicles[df_vehicles['VEHICLE_TYPE'].isin(top_vehicle_types)], order=top_vehicle_types)
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Distribution of Top 10 Vehicle Types')
plt.xticks(rotation=45)
plt.show()

# Plot the distribution of contributing factors
plt.figure(figsize=(12, 6))
sns.countplot(x='CONTRIBUTING_FACTOR_1', data=df_vehicles, order=df_vehicles['CONTRIBUTING_FACTOR_1'].value_counts().index[:10])
plt.xlabel('Contributing Factor 1')
plt.ylabel('Count')
plt.title('Distribution of Contributing Factors (Top 10)')
plt.xticks(rotation=45)
plt.show()

# Plot the correlation heatmap of numerical columns
plt.figure(figsize=(10, 8))
sns.heatmap(df_vehicles.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned vehicles dataset from the CSV file
df_vehicles = pd.read_csv('updated_cleaned_vehicles_data.csv')

# Handle missing values in 'VEHICLE_TYPE'
df_vehicles['VEHICLE_TYPE'].fillna('Unknown', inplace=True)

# Standardize inconsistent values in 'CONTRIBUTING_FACTOR_1'
df_vehicles['CONTRIBUTING_FACTOR_1'].replace(['1', '80'], 'Unknown', inplace=True)

# Standardize inconsistent values in 'VEHICLE_TYPE'
df_vehicles['VEHICLE_TYPE'] = df_vehicles['VEHICLE_TYPE'].str.upper()
df_vehicles['VEHICLE_TYPE'].replace(['10 PAASENG', 'COMM TRAIL'], ['10 PASSENGER', 'COMMERCIAL TRAILER'], inplace=True)

# Save the final cleaned dataset to a new CSV file
df_vehicles.to_csv('final_cleaned_vehicles_data.csv', index=False)

# Plot the distribution of top vehicle types
top_vehicle_types = df_vehicles['VEHICLE_TYPE'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.countplot(x='VEHICLE_TYPE', data=df_vehicles[df_vehicles['VEHICLE_TYPE'].isin(top_vehicle_types)], order=top_vehicle_types)
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Distribution of Top 10 Vehicle Types')
plt.xticks(rotation=45)
plt.show()

# Plot the distribution of contributing factors
plt.figure(figsize=(12, 6))
sns.countplot(x='CONTRIBUTING_FACTOR_1', data=df_vehicles, order=df_vehicles['CONTRIBUTING_FACTOR_1'].value_counts().index[:10])
plt.xlabel('Contributing Factor 1')
plt.ylabel('Count')
plt.title('Distribution of Contributing Factors (Top 10)')
plt.xticks(rotation=45)
plt.show()

# Plot the correlation heatmap of numerical columns
plt.figure(figsize=(10, 8))
sns.heatmap(df_vehicles.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Numerical Columns')
plt.show()
```
